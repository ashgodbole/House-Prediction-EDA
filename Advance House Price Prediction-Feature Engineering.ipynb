```python
# Here we will check the percentage of nan values present in each feature
#1- step make the list of features which has missing values

feature_nan =[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in feature_nan:
    print("{}:{}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

### replace missing values with a new label

```python
def replace_cat_feature(dataset,feature_nan):
    data=dataset.copy()
    data[feature_nan]=data[feature_nan].fillna('missing')
    return data

dataset=replace_cat_feature(dataset,feature_nan)

dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

### search the numarical nan value 

```python
numarical_with_nan =[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']

for feature in numarical_with_nan:
    print("{}:{}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

### replacing the numarical vvalues

```python
for feature in numarical_with_nan:
    ## we will replace by using median since there are outliers
    median_value=dataset[feature].median()
    
    ## create a new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numarical_with_nan].isnull().sum()
```

### Temporal variables (Date Time Variables)

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
##numarical value outliers are consider for this process 

## convert skew distribution into log normal distribution

import numpy as np
num_features=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for feature in num_features:
    dataset[feature]=np.log(dataset[feature])
```

### handling Rare Categorical Feature

we will remove categorial variable that are present less than 1% of the observations

```python
categorial_features=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
```

```python
categorial_features
```

```python
for feature in categorial_features:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head()
```

```python
for feature in categorial_features:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```
